```
# importing python libarary

import pandas as pd 

import numpy as np
```

```
#loading our dataset

Autolib_datset = 'http://bit.ly/Autolibdataseturl' 
A_df = pd.read_csv('/content/Autolib_dataset (2).csv')

#Previewing our dataset

A_df .head()
```

```
#Dropping irrelevant columns from the dataset

A_df .drop(columns = ['Address','Cars','Charge Slots', 'Charging Status', 'Displayed comment', 'Geo point', 'Scheduled at', 'Slots', 'Station type', 'Status', 'Subscription status','Rental status','Station type'], axis = 1, inplace =True)
A_df .head()
```

```
#Merging dates  column

A_df ['Date'] =  A_df['day'].map(str) + '/' + A_df['month'].map(str) + '/' + A_df['year'].map(str)

#converting date into a standard format

A_df['Date'] = pd.to_datetime(A_df['Date'])

#add datetime colunm in the standard format

A_df['datetime'] = A_df['day'].map(str) + '/' + A_df['month'].map(str) + '/' + A_df['year'].map(str) + ' ' + A_df['hour'].map(str) + ':' + A_df['minute'].map(str) 
#add datetime colunm in the standard format

A_df['datetime'] = pd.to_datetime(A_df['datetime'])
A_df
```

```
#Dropping null values
A_df.dropna()

	
```

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
# #filtering out other cities to just have paris City alone ordered by datetime.

a_df= A_df[(A_df['Kind'] == "STATION")]
a_df[(a_df['City'] == "Paris")].groupby(['ID', 'Bluecar counter']).first().sort_values(by = 'datetime')
```

##Identify the popular hour of the day that is popular for picking up shared electric car in the city of Paris in the month of April.

```
#To find out the popular hour for picking up cars we selected the hour with the most negative cumulative values since -ve values means picking
a_df['diff'] =  a_df['Bluecar counter'].diff()
a_df[(a_df['diff'] < 0)].groupby(['hour'])['diff'].sum().sort_values(ascending =True)[:1]
```

***popular hour for picking is 21:00 hrs***





##What is the most popular hour for returning cars?

```
#To find out the popular hour for picking up cars we selected the hour with the most +ve cumulative values since +ve values means returning
a_df[(a_df['diff'] > 0)].groupby(['hour'])['diff'].sum().sort_values(ascending =False)[:1]
```

***popular hour for returning is 06:00 hrs***
